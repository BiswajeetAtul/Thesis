```python
cwd = os.getcwd()
model_runtime_folder = cwd +"\\logs\\fit\\runtime"
model_history_folder = cwd + "\\logs\\hist"
print(model_runtime_folder)
print(model_history_folder)
```

```python
# creating a dictionary with all evaluation metrics csv files
model_eval_metrics={}
model_runtimes_ ={}
# adding all the csv files in the directory to the
for  root, dir,file in  os.walk(model_history_folder):
    for item in file:
        if '.csv' in item:
            print(item)
            model_eval_metrics[item[:-4]]= pd.read_csv(model_history_folder+"\\"+item)
# print(model_eval_metrics)

# adding all the csv files in the directory to the
for  root, dir,file in  os.walk(model_runtime_folder):
    for item in file:
        if '.csv' in item:
            print(item)
            model_runtimes_[item[:-4]]= pd.read_csv(model_runtime_folder+"\\"+item)

print("Files Read :")
print("======================================================")
print(model_eval_metrics.keys())
print("======================================================")
print(model_runtimes_.keys())
```

```python
for df in model_eval_metrics.keys():
    model_eval_metrics[df]["epocs"]=model_eval_metrics[df]["Unnamed: 0"]+1
    model_eval_metrics[df].columns=['Unnamed: 0', 'loss', 'acc', 'categorical_accuracy','top_k_categorical_accuracy', 'precision', 'recall', 'true_positives','true_negatives', 'false_positives', 'false_negatives', 'val_loss','val_acc', 'val_categorical_accuracy', 'val_top_k_categorical_accuracy','val_precision', 'val_recall', 'val_true_positives','val_true_negatives', 'val_false_positives', 'val_false_negatives','lr',"epocs"]

    display(model_eval_metrics[df].head(3))


for df in model_runtimes_.keys():
    display(model_runtimes_[df])
```

```python
def generate_runtime_graph(df_dict):
    """
    df_dict = dictionary containing all the file names
    Plots the total time take for the model to run 20 epocs in training
    """
    for df_name in df_dict.keys():
        _df_train = df_dict[df_name][df_dict[df_name]["phase"]=="train"]
        _df_test = df_dict[df_name][df_dict[df_name]["phase"]=="test"]

        # print(_df_test.shape)
        # print(_df_train.shape)
        # plt.figure(figsize=(10,10))
        # plt.plot(_df_train["model"], _df_train["total_time"],  '.-',label="total_time")
        # plt.plot(_df_test["model"], _df_test["total_time"],  '.-',label="total_time")
        # plt.legend()

        # create figure and axis objects with subplots()
        fig,ax = plt.subplots(sharey=True)
        fig.set_figwidth(20)
        fig.set_figheight(10)
        # make a plot
        ax.plot(_df_train["model"], _df_train["total_time"], color="red", marker="o", label="total_train_time", linewidth=2, markersize=12)
        # set x-axis label
        ax.set_xlabel("Model",fontsize=14)
        # set y-axis label
        ax.set_ylabel("Train Time",color="red",fontsize=14)
        ax.legend(loc='upper center', bbox_to_anchor=(0.5,1))

        # twin object for two different y-axis on the sample plot
        ax2=ax.twinx()
        # make a plot with different y-axis using second axis object
        ax2.plot(_df_test["model"],  _df_test["total_time"],color="blue",marker="s" ,label="total_test_time", linestyle='--',linewidth=2, markersize=12)
        ax2.set_ylabel("Test TIme",color="blue",fontsize=14)
        ax2.legend(loc='upper center', bbox_to_anchor=(0.5,0.95))

        fig.suptitle(df_name)
        plt.show()
    # sns.lineplot(data=df_dict[df_name], x="HiddenNodes",y="term_wise_accuracy", label="term_wise_accuracy")
```

```python
generate_runtime_graph(model_runtimes_)
```

```python
def generate_model_performance_graphs(df_dict):
    """
    df_dict = dictionary containing all the file names
    plots graph for showing how the metrics raise each epoch
    """
    for df_name in df_dict.keys():

        # plt.figure(figsize=(20,10))
        # plt.title(df_name)
        _df= df_dict[df_name]
        fig, ax= plt.subplots(2,1 , sharex=True, sharey=True)
        fig.set_figwidth(20)
        fig.set_figheight(10)

        ax[0].title.set_text("Training")
        # ax[0,0].plot(_df["HiddenNodes"], _df["Metrics"])
        # ax[0].plot(_df["epocs"], _df["loss"], '.-', label="loss")
        ax[0].plot(_df["epocs"], _df["categorical_accuracy"],  '.-',label="categorical_accuracy")
        ax[0].plot(_df["epocs"], _df["top_k_categorical_accuracy"],  '.-',label="top_k_categorical_accuracy")
        ax[0].legend()


        ax[1].title.set_text("Validation")

        # ax[1].plot(_df["epocs"], _df["val_loss"],  '.-',label="val_loss")
        ax[1].plot(_df["epocs"], _df["val_categorical_accuracy"],  '.-',label="val_categorical_accuracy")
        ax[1].plot(_df["epocs"], _df["val_top_k_categorical_accuracy"],  '.-',label="val_top_k_categorical_accuracy")
        ax[1].legend()


        # ax[1,0].title.set_text("Macro Metrics")

        # ax[1,0].plot(_df["epocs"], _df["sklearn_macro_precision"], '.-', label="sklearn_macro_precision")
        # ax[1,0].plot(_df["epocs"], _df["sklearn_macro_recall"],  '.-',label="sklearn_macro_recall")
        # ax[1,0].plot(_df["epocs"], _df["sklearn_macro_f1"],  '.-',label="sklearn_macro_f1")
        # ax[1,0].legend()

        # ax[1,1].title.set_text("Micro Metrics")

        # ax[1,1].plot(_df["epocs"], _df["sklearn_micro_precision"], '.-', label="sklearn_micro_precision")
        # ax[1,1].plot(_df["epocs"], _df["sklearn_micro_recall"], '.-', label="sklearn_micro_recall")
        # ax[1,1].plot(_df["epocs"], _df["sklearn_micro_f1"], '.-', label="sklearn_micro_f1")
        # ax[1,1].legend()


        fig.suptitle(df_name)
```
