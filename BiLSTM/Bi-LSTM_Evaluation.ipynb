```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
# getting the current working directory and and ealuation metrics folder
cwd = os.getcwd()
evaluation_folder = "\\logs\\fit\\eval"
runtime_folder = "\\logs\\fit\\runtime"
evaluatio_metrics_folder =cwd +evaluation_folder
print(evaluatio_metrics_folder)
runtime_folder =cwd +runtime_folder
print(runtime_folder)
```

```python
# creating a dictionary with all evaluation metrics csv files
model_eval_metrics={}

# adding all the csv files in the directory to the
for  root, dir,file in  os.walk(evaluatio_metrics_folder):
    for item in file:
        if '.csv' in item:
            print(item)
            model_eval_metrics[item[:-4]]= pd.read_csv(evaluatio_metrics_folder+"\\"+item)
# print(model_eval_metrics)
```

```python
# creating a dictionary with all evaluation metrics csv files
runtime_metrics={}

# adding all the csv files in the directory to the
for  root, dir,file in  os.walk(runtime_folder):
    for item in file:
        if '.csv' in item:
            print(item)
            runtime_metrics[item[:-4]]= pd.read_csv(runtime_folder+"\\"+item)
# print(model_eval_metrics)
```

```python
# function to show the evaluation metrics across thresholds

#sns.set_theme() 
def showEvalMetricsInThresholds(df_dict, embedding_type=''):

    subset=[]
    
    for df_name in df_dict.keys():
        if embedding_type in df_name:
            subset.append(df_name)
    
    for df_name in subset:
        # print(df_name)
        _df = df_dict[df_name]
        fig, ax= plt.subplots(3,1, sharex=True, sharey=False)
        fig.set_figwidth(20)
        fig.set_figheight(10)

        ax[0].title.set_text("Accuracy, Term wise Accuracy and Hamming Loss")
        # ax[0,0].plot(_df["HiddenNodes"], _df["Metrics"])
        ax[0].plot(_df["threshold"], _df["hamming_loss"], '.-', label="hamming_loss")
        ax[0].plot(_df["threshold"], _df["term_wise_accuracy"],  '.-',label="term_wise_accuracy")
        ax[0].plot(_df["threshold"], _df["accuracy"],  '.-',label="sklearn_accuracy")
        ax[0].legend()


        ax[1].title.set_text("Macro Metrics")

        ax[1].plot(_df["threshold"], _df["macro_precision"], '.-', label="sklearn_macro_precision")
        ax[1].plot(_df["threshold"], _df["macro_recall"],  '.-',label="sklearn_macro_recall")
        ax[1].plot(_df["threshold"], _df["macro_f1"],  '.-',label="sklearn_macro_f1")
        ax[1].legend()

        ax[2].title.set_text("Micro Metrics")

        ax[2].plot(_df["threshold"], _df["micro_precision"], '.-', label="sklearn_micro_precision")
        ax[2].plot(_df["threshold"], _df["micro_recall"], '.-', label="sklearn_micro_recall")
        ax[2].plot(_df["threshold"], _df["micro_f1"], '.-', label="sklearn_micro_f1")
        ax[2].legend()


        fig.suptitle(df_name)
        fig.savefig(cwd+'\\logs\\images\\'+df_name+'.png')

```

```python
showEvalMetricsInThresholds(model_eval_metrics, "xlnet_t2")
```

```python
sns.set_theme() 
def showRuntimePlots(df_dict):
    mapper ={
        "bert_t1" : "BERT_StopWords",
        "bert_t2" : "BERT_NoStopWords",
        "xlnet_t1" : "XLNet_StopWords",
        "xlnet_t2" : "XLNet_NoStopWords"
    }

    for df_name in df_dict:
        # print(df_name)
        _df = df_dict[df_name].copy()
        _df.model = _df.model.apply(lambda x: x[17:]).map(mapper)
        _df_train = _df[_df["phase"]=="train"]
        _df_test = _df[_df["phase"]=="test"]
        width = 0.20

        plt.figure(figsize=(5,5))
        plt.bar(_df_train["model"],_df_train["total_time"],width, color='b', label="Train Time")
        plt.bar(_df_test["model"],_df_test["total_time"],width, color='g', label="Test Time")
        plt.xticks(rotation=10)
        plt.ylim(0,120)
        plt.legend(loc =2)
        plt.title(df_name+ " @ 20 Epochs")
        plt.savefig(cwd+'\\logs\\images\\'+df_name+'.png')
```

```python
showRuntimePlots(runtime_metrics)
```

```python
# function to show the evaluation metrics across thresholds

sns.set_theme() 
def showEvalMetricsInThresholds_alt(df_dict, embedding_type=''):

    subset=[]
    
    for df_name in df_dict.keys():
        if embedding_type in df_name:
            subset.append(df_name)

    fig, ax= plt.subplots(4,1, sharex=True, sharey=False)
    fig.set_figwidth(20)
    fig.set_figheight(10)
    count =0 
    color = ['b','g','r','c','m']  
    for df_name in subset:
        # print(df_name)
        _df = df_dict[df_name]


        ax[0].title.set_text("micro_precision")
        ax[0].plot(_df["threshold"], _df["micro_precision"], 's-', c=color[count], label=df_name+"sklearn_micro_precision")
        # # ax[0,0].plot(_df["HiddenNodes"], _df["Metrics"])
        # ax[0].plot(_df["threshold"], _df["hamming_loss"], '.-', label="hamming_loss")
        # ax[0].plot(_df["threshold"], _df["term_wise_accuracy"],  '.-',label="term_wise_accuracy")
        # 
        ax[0].legend()


        ax[1].title.set_text("micro_recall")
        ax[1].plot(_df["threshold"], _df["micro_recall"], '*-',c=color[count], label=df_name+"sklearn_micro_recall")

        # ax[1].plot(_df["threshold"], _df["macro_precision"], '.-', label="sklearn_macro_precision")
        # ax[1].plot(_df["threshold"], _df["macro_recall"],  '.-',label="sklearn_macro_recall")
        # ax[1].plot(_df["threshold"], _df["macro_f1"],  '.-',label="sklearn_macro_f1")
        ax[1].legend()

        ax[2].title.set_text("micro_f1")
        ax[2].plot(_df["threshold"], _df["micro_f1"], '.-', c= color[count] ,label=df_name+"sklearn_micro_f1")
        ax[2].legend()

        ax[3].title.set_text("accuracy") 
        ax[3].plot(_df["threshold"], _df["accuracy"], '+-', c= color[count],label="sklearn_accuracy")
        ax[3].legend()
        count+=1
        # fig.suptitle(df_name)
        # fig.savefig(cwd+'\\logs\\images\\'+df_name+'.png')

```
